### Pandas melt() used to convert data to tidy format

```python
# Example 1
df = pd.DataFrame({"Q1": [10, 11, 12],
                 "Q2": [25, 27, 29],
                 "Q3": [33, 38, 39],
                  "Q4": [33, 38, 39]})
df
```

### Pandas Melt

```python
df_tidy = df.melt()
df_tidy
```

```python
## tidy() and rename columns
df_tidy = df.melt().rename(columns={'variable':'Quarter', 'value':'Qty_Sold'})
df_tidy
```

```python
# Example 2 

df = pd.DataFrame({'City': ['Houston', 'Austin', 'Hoover'],
                   'State': ['Texas', 'Texas', 'Alabama'],
                   'Mango':[4, 10, 90],
                   'Orange': [10, 8, 14], 
                   'Watermelon':[40, 99, 43]}
                 )
df
```

```python
df.melt(id_vars = ['City', 'State'],
           value_vars = ['Mango', 'Orange', 'Watermelon'],
            var_name='Fruit', value_name='Qty_Sold')

# id_vars : specifies which column(s) you want to leave untouched.
```

```python
### Example 3

data_url = "https://goo.gl/ioc2Td"
gapminder = pd.read_csv(data_url)
gapminder.head()
```

The gapminder data in wide form contains lifeExp, gdpPercap and population per year as columns. 

```python
gapminder.columns
```

```python
df_gdp = gapminder.loc[:, gapminder.columns.str.contains('^gdpPercap|^co')]
```

```python
df_gdp.head(3)
```

```python
df_gdp_tidy = df_gdp.melt(id_vars=['continent', 'country'], var_name='Year', value_name='gdpPercap')
```

```python
df_gdp_tidy.head(3)
```

```python
df_gdp_tidy['Year'] = df_gdp_tidy['Year'].str.replace('gdpPercap_','').astype('int')
```

```python
df_gdp_tidy.head(3)
```

```python
# Convert LifeExp into tidy dtaframe

df_lifeExp = gapminder.loc[:, gapminder.columns.str.contains('^lifeExp|^co')]

df_lifeExp_tidy = df_lifeExp.melt(id_vars = ['continent', 'country'],
                                 var_name='Year', value_name='lifeExp')

df_lifeExp_tidy['Year'] = df_lifeExp_tidy['Year'].str.replace('lifeExp_','').astype('int')

df_lifeExp_tidy.head(3)
```

```python
# population columns  into tidy dataframe

df_pop = gapminder.loc[:, gapminder.columns.str.contains('^pop|^co')]

df_pop_tidy = df_pop.melt(id_vars = ['continent', 'country'],
                                 var_name='Year', value_name='Population')
df_pop_tidy['Year'] = df_pop_tidy['Year'].str.replace('pop_','').astype('int')
df_pop_tidy.head()
```

```python
df_final = pd.merge(left=df_gdp_tidy, right=df_lifeExp_tidy, on=['continent', 'country', 'Year']).\
            merge(right=df_pop_tidy, on=['continent', 'country', 'Year'])
df_final.head()
```
